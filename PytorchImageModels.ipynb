# Create a Model

```python
import timm
import torch

model = timm.create_model('resnet34')
x = torch.randn(1,3,224,224)
model(x).shape
```

* create_model is a factory method, that can be used to create over 300 models as a part of the timm library

* To create a pretrained model, simply pass **pretrained = True** when creating the model

```python
pretrained_resnet_34 = timm.create_model('resnet34', pretrained=True)
```

* to create a model with custom number of classes, pass **num_classes = <number of classes>** when creating the model

```python
import timm
import torch

pretrained_resnet_34_with_classes = timm.create_model('resnet34', num_classes=10)
x = torch.randn(1,3,224,224)
pretrained_resnet_34_with_classes(x).shape
```
